This notebook develops a first-order thermal model for plasma-facing components
(PFCs) subjected to transient fusion heat loads. The purpose is to estimate
surface temperature evolution and corresponding thermal radiation that would be
observed by the proposed FusionCAM diagnostic system.

Conceptual design study

## Imports

```python
from pathlib import Path
import sys

project_root = Path.cwd().parent
sys.path.append(str(project_root))

from physics.heat_flux_model import demo_heat_pulse

import numpy as np
import matplotlib.pyplot as plt
plt.style.use("default")
```

## Motivation

Plasma-facing components in tokamaks routinely experience heat loads ranging from
approximately 0.5–20 MW/m² depending on plasma conditions.

A thermal imaging diagnostic must estimate

- transient surface temperature
- radiated power
- dynamic range
- frame rate requirements

before hardware can be designed.

This notebook develops a simplified lumped thermal model to estimate these
quantities.

## Heat Pulse

```python
duration = 5  # seconds
data = demo_heat_pulse(duration)
```

```python
t = data["t"]
q_in = data["q_in"]
T = data["T"]
q_rad = data["q_rad"]
q_cond = data["q_cond"]
```

```python
plt.figure(figsize=(9,5))

plt.plot(t, q_in/1e6, label="Plasma Heat Flux")
plt.plot(t, q_rad/1e6, label="Radiative Loss")
plt.plot(t, q_cond/1e6, label="Conduction to Bulk")
plt.xlabel("Time (s)")
plt.ylabel("Heat Flux (MW/m²)")
plt.title(f"Heat Balance During a {duration}-Second Plasma Pulse")

plt.grid(True)
plt.legend()
plt.savefig(f"../figures/heat_balance_during_{duration}-sec_plasma_pulse.png",dpi='figure', bbox_inches='tight')
plt.show()
```
